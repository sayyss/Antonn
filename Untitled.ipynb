```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import time
from pymongo import MongoClient
import datetime
```

```python
client = MongoClient()
db = client.DiscordServers
servers = db.servers
```

```python
import db_commands
db = db_commands.DB()
```

```python
data = db.getDailyMsgs(710551528301395988)
```

```python
x = []
y = []

for i in data:

    time = i['time']
    time = datetime.datetime.fromtimestamp(time)
    time = time.time()
    time = time.strftime("%H:%M")

    x.append(time)
    y.append(i['count'])
     
plt.figure(figsize=(17,5))
plt.plot(x,y)
```
